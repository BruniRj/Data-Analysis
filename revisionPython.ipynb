```python
def temperature_advice(temp):
    if temp > 30 :
        return "il fait chaud "
    elif temp >= 15 : 
        return "le temps est agreable"
    else :
        return "prend un manteau "

print(temperature_advice(14))
```

```python
# calcul de l'imc 
poid = 10
taille = 1.75
imc = poid / ( taille ** 2 )
if imc < 18.5:
    print("Maigreur")
elif imc < 25:
    print("Normal")
else:
    print("Surpoids")
```

```python
# Dataset : notes de 5 étudiants [14, 8, 16, 11, 9]
notes = [14, 8, 16, 11, 9 ]
# 1. Calcule la moyenne
# 2. Affiche combien ont la moyenne (>=10)
# 3. Affiche la meilleure note
# 4. Transforme en dict {"Jean":14, "Sofia":8, ...} (invente prénoms)

# 1 : calcule de la moyenne 
#  somme des valeurs / nombres des valeurs
moyenne = sum(notes) / len(notes)
print(f"voici la moyenne : {moyenne:.2f}")
print(f"voici la moyenne {moyenne}")

# 2 affichage sur combien ont la moyenne 
moyenne_etudiant = 0
for note in notes:
    if note >= 10 : 
        moyenne_etudiant+=1
print(f"voici les etudiants qui ont de la moyenne : {moyenne_etudiant}")

#  meilleur note : 
meilleur = max(notes)
mauvais = min(notes)
print(f"voici l'etudiant ayant la note maximale : {meilleur}  et voici le mauvais {mauvais}")

# transformation du liste en dict
etudiants = ["Jean", "Sofia", "Ali", "Maria", "David"]
notes_dict = dict(zip(etudiants, notes))
print(f"voici les dict par etudiants : {notes_dict}")
```
